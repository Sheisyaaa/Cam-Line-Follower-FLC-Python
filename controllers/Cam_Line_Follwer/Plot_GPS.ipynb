```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_gps_data():
    # Membaca data dari file
    file_path = 'gps_data.csv'
    data = pd.read_csv(file_path)
    
    # Plotting tracking gerak robot
    plt.figure(figsize=(10, 5))
    plt.plot(data['gps value x'], data['gps value y'], label='Track', marker='o', linestyle='-', color='b')  # Gunakan marker untuk menandai posisi
    
    plt.title('Tracking Gerak Robot Selama Simulasi')
    plt.xlabel('GPS X')
    plt.ylabel('GPS Y')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')  # Pastikan skala sumbu x dan y sama untuk representasi yang akurat
    plt.show()

# Memanggil fungsi untuk memplot data
plot_gps_data()
```
